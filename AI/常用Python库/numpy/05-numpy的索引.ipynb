ğŸŒŸ é‡ç‚¹æç¤ºï¼šåˆ‡ç‰‡å’Œç´¢å¼•æ˜¯åœ¨ç°æœ‰æ•°ç»„ä¸Šæ“ä½œä»¥è·å–æ‰€éœ€ã€Œéƒ¨åˆ†ã€å…ƒç´ çš„æ–¹æ³•ã€‚å…¶æ ¸å¿ƒæ˜¯æ ¹æ® `start:stop:step` æ¨¡å¼æŒ‰ç»´åº¦æ“ä½œæ•°ç»„ã€‚

### æ‹¼æ¥

åœ¨æ•°æ®å¤„ç†ä¸­ï¼Œæˆ‘ä»¬ç»å¸¸éœ€è¦å°†å¤šä¸ªæ•°ç»„åˆå¹¶æˆä¸€ä¸ªæ›´å¤§çš„æ•°ç»„ã€‚è¿™å¯ä»¥é€šè¿‡æ‹¼æ¥æˆ–å †å å®Œæˆã€‚NumPy æä¾›äº†å¤šç§æ–¹æ³•æ¥å®ç°è¿™ä¸€ç‚¹ï¼Œæˆ‘ä»¬å°†é‡ç‚¹ä»‹ç» `np.concatenate` å’Œ `np.stack`ï¼Œè¿™ä¸¤ä¸ªå‡½æ•°æä¾›äº†åˆå¹¶æ•°ç»„çš„åŸºæœ¬åŠŸèƒ½ã€‚

#### `np.concatenate`

`np.concatenate` æ˜¯ NumPy ä¸­æœ€å¸¸ç”¨çš„æ•°ç»„æ‹¼æ¥å‡½æ•°ä¹‹ä¸€ã€‚å®ƒå¯ä»¥æ²¿ç€æŒ‡å®šçš„è½´è¿æ¥æ•°ç»„åºåˆ—ã€‚

```python
# åˆ›å»ºä¸¤ä¸ªå½¢çŠ¶ç›¸åŒçš„æ•°ç»„
arr1 = rng.random((2, 3))
arr2 = rng.random((2, 3))

# é»˜è®¤æƒ…å†µä¸‹ï¼Œ`np.concatenate` æ²¿ç€ç¬¬ä¸€ä¸ªè½´ï¼ˆaxis=0ï¼Œå³è¡Œï¼‰è¿›è¡Œè¿æ¥
concatenated_arr = np.concatenate((arr1, arr2))
print(concatenated_arr.shape)
```

```python
# å¯ä»¥æŒ‡å®š `axis` å‚æ•°æ¥æ²¿ç€ä¸åŒçš„è½´æ‹¼æ¥æ•°ç»„ï¼Œè¿™é‡Œæ˜¯æ²¿ç€åˆ—ï¼ˆaxis=1ï¼‰
concatenated_arr_axis1 = np.concatenate((arr1, arr2), axis=1)
print(concatenated_arr_axis1.shape)
```

è¿™å°†è¾“å‡ºä¸€ä¸ªæ–°æ•°ç»„ï¼Œå…¶ä¸­ç¬¬äºŒä¸ªæ•°ç»„çš„åˆ—è¢«æ·»åŠ åˆ°ç¬¬ä¸€ä¸ªæ•°ç»„çš„å³ä¾§ã€‚

#### `np.stack`

`np.stack` æ˜¯å¦ä¸€ä¸ªç”¨äºå †å æ•°ç»„çš„å‡½æ•°ï¼Œä¸åŒäº `np.concatenate`ï¼Œ`np.stack` ä¼šåˆ›å»ºä¸€ä¸ªæ–°çš„è½´ã€‚

```python
# ä½¿ç”¨ `np.stack` æ¥å †å ä¸¤ä¸ªæ•°ç»„ï¼Œå®ƒä¼šåœ¨ç»“æœä¸­æ·»åŠ ä¸€ä¸ªæ–°çš„è½´
stacked_arr = np.stack((arr1, arr2))
print(stacked_arr)
print("Shape of stacked array:", stacked_arr.shape)
```

```python
# å¯ä»¥æŒ‡å®š `axis` å‚æ•°æ¥æ²¿ç€ä¸åŒçš„è½´å †å æ•°ç»„ï¼Œè¿™é‡Œæˆ‘ä»¬æ²¿ç€æœ€å†…ä¾§çš„è½´ï¼ˆaxis=2ï¼‰è¿›è¡Œå †å 
stacked_arr_axis2 = np.stack((arr1, arr2), axis=2)
print(stacked_arr_axis2)
print("Shape of stacked array along axis 2:", stacked_arr_axis2.shape)
```

### é‡å¤

æœ‰æ—¶æˆ‘ä»¬éœ€è¦å°†æ•°ç»„ä¸­çš„å…ƒç´ æ²¿ç€æŒ‡å®šçš„è½´é‡å¤æŸäº›æ¬¡æ•°ï¼Œ`np.repeat` å‡½æ•°æ­£æ˜¯ä¸ºæ­¤è®¾è®¡çš„ã€‚

```python
# åˆ›å»ºä¸€ä¸ª 3x4 çš„éšæœºæ•´æ•°æ•°ç»„
arr = rng.integers(0, 10, (3, 4))
print(arr)
```

```python
# æ²¿ç€ axis=0 é‡å¤æ¯ä¸€è¡Œä¸¤æ¬¡
repeated_arr_axis0 = np.repeat(arr, 2, axis=0)
print(repeated_arr_axis0)
```

```python
# æ²¿ç€ axis=1 é‡å¤æ¯ä¸€åˆ—ä¸‰æ¬¡
repeated_arr_axis1 = np.repeat(arr, 3, axis=1)
print(repeated_arr_axis1)
```

### åˆ†æ‹†

ä¸æ‹¼æ¥ç›¸åï¼Œåˆ†æ‹†æ˜¯å°†ä¸€ä¸ªå¤§æ•°ç»„åˆ†å‰²æˆå¤šä¸ªå°æ•°ç»„çš„è¿‡ç¨‹ã€‚`np.split` æä¾›äº†ä¸€ä¸ªé€šç”¨çš„åˆ†å‰²æ–¹æ³•ã€‚

```python
# åˆ›å»ºä¸€ä¸ª 6x4 çš„éšæœºæ•´æ•°æ•°ç»„
arr = rng.integers(1, 100, (6, 4))
print(arr)
```

```python
# é»˜è®¤æƒ…å†µä¸‹ï¼Œ`np.split` æ²¿ç€ç¬¬ä¸€ä¸ªè½´ï¼ˆaxis=0ï¼‰åˆ†å‰²æ•°ç»„ï¼Œè¿™é‡Œæˆ‘ä»¬å°†å…¶åˆ†å‰²æˆ 3 ä¸ªç›¸åŒå¤§å°çš„å°æ•°ç»„
split_arr = np.split(arr, 3)
print(split_arr)
```

```python
# æˆ‘ä»¬ä¹Ÿå¯ä»¥æ²¿ç€åˆ—ï¼ˆaxis=1ï¼‰åˆ†å‰²æ•°ç»„ï¼Œè¿™é‡Œæˆ‘ä»¬å°†å…¶åˆ†å‰²æˆ 2 ä¸ªç›¸åŒå¤§å°çš„å°æ•°ç»„
split_arr_axis1 = np.split(arr, 2, axis=1)
print(split_arr_axis1)
```

### æ¡ä»¶ç­›é€‰

åœ¨å¤„ç†æ•°ç»„æ—¶ï¼Œæˆ‘ä»¬ç»å¸¸æ ¹æ®æ¡ä»¶é€‰æ‹©æˆ–ä¿®æ”¹å…ƒç´ ã€‚`np.where` æ˜¯ä¸€ä¸ªéå¸¸æœ‰ç”¨çš„å‡½æ•°ï¼Œå®ƒè¿”å›æ»¡è¶³æ¡ä»¶çš„å…ƒç´ çš„ç´¢å¼•ã€‚

```python
# åˆ›å»ºä¸€ä¸ª 6x4 çš„éšæœºæ•´æ•°æ•°ç»„
arr = rng.integers(1, 100, (6, 4))
print(arr)
```

```python
# ä½¿ç”¨æ¡ä»¶ç­›é€‰æ¥åˆ›å»ºä¸€ä¸ªå¸ƒå°”æ•°ç»„
condition = arr > 50
print(condition)
```

```python
# ä½¿ç”¨ `np.where` æ¥æ‰¾åˆ°æ»¡è¶³æ¡ä»¶çš„å…ƒç´ çš„ç´¢å¼•
indices = np.where(arr > 50)
print(indices)
```

```python
# ä½¿ç”¨ `np.where` è¿›è¡Œæ¡ä»¶èµ‹å€¼ï¼Œå°†æ‰€æœ‰å°äºç­‰äº 50 çš„å…ƒç´ æ›¿æ¢ä¸º -1
new_arr = np.where(arr > 50, arr, -1)
print(new_arr)
```
